#### Load terms

```python
sent_file = open("AFINN-111.txt")
```

```python
scores = {} # initialize an empty dictionary
```

```python
for line in sent_file:
  term, score  = line.split("\t")  # The file is tab-delimited. "\t" means "tab character"
  scores[term] = int(score)  # Convert the score to an integer.
```

#### Loads tweets from .json

```python
import json
import re
```

```python
tweet_file = open('output.txt')
```

```python
tweets = []
```

```python
for line in tweet_file:
    dic = json.loads(line)
    try:
        tweets.append(dic['text'])
    except:
        pass # no real tweet
```

#### Q2 and Q3

```python
new_terms = {}
```

```python
for tweet in tweets[:5]:
    score = 0
    words = re.findall('[a-zA-Z]+', tweet)
    words = [word.lower().strip() for word in words]
    for word in words:
        try:
           score = score + scores[word]
        except:
            # Q3: if word is not on loaded terms
            try:
                new_terms[word]
                # word already on new_terms
            except:
                # word is not on new_terms
                new_terms[word] = []
    #print score, tweet
    
    # Q3: assign score to all words
    for word in words:
        try:
            new_terms[word].append(score)
        except:
            pass
            # word is on scores (not new term) so is not needed
```

#### Q3: print

```python
for key in new_terms.iterkeys():
    scores = new_terms[key]
    print key, reduce(lambda x, y: x + y, scores) / len(scores)
```

#### Q4

```python
frequency = {}
```

```python
for tweet in tweets[:100]:
    words = re.findall('[a-zA-Z]+', tweet)
    words = [word.lower().strip() for word in words]
    for word in set(words):
        try:
            frequency[word] = frequency[word] + 1
        except:
            frequency[word] = 1
```

```python
total_occurrences = reduce(lambda x, y: x + y, frequency.values())
```

```python
from __future__ import division
```

```python
for key in frequency.iterkeys():
    print key, frequency[key] / total_occurrences
```

#### Q6

```python
import json
```

```python
tweet_file = open('output.txt')
```

```python
hashtags_count = {}
```

```python
for line in tweet_file:
    dic = json.loads(line)
    try:
        hashtags = dic['entities']['hashtags']
    except:
        pass # no real tweet or no hashtags
    for hashtag in hashtags:
        hashtag = hashtag['text'].lower()
        try:
           hashtags_count[hashtag] = hashtags_count[hashtag] + 1
        except:
            hashtags_count[hashtag] = 1
```

```python
sorted_tags = sorted(hashtags_count, key=hashtags_count.get, reverse=True)
```

```python
for tag in sorted_tags[:10]:
    print tag, hashtags_count[tag]
```
